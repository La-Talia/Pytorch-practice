Reshape and view

```python
my_torch = my_torch.reshape(2,5)
print(my_torch)
```

```python
my_torch = my_torch.reshape(-1,5)
print(my_torch)
```

```python
my_torch2 = my_torch.view(5,-1)
print(my_torch2)
```

```python
my_torch[1] = 987
print(my_torch)
print(my_torch2)
```

Tensor Math operations

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
#addition
tensor_a + tensor_b
```

```python
#addition long hand
torch.add(tensor_a,tensor_b)
```

```python
#subtration
tensor_a - tensor_b
```

```python
#sub function
torch.sub(tensor_a, tensor_b)
```

```python
#multiply
tensor_a * tensor_b
```

```python
#mul func
torch.mul(tensor_a, tensor_b)
```

```python
#div
tensor_a / tensor_b
```

```python
#division
torch.div(tensor_b, tensor_a)
```

```python
#remainders
tensor_b % tensor_a
```
